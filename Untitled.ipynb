## Checking relationship of Year with Price

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Set plot style and aesthetics
sns.set(style="whitegrid", context="notebook", palette="deep")

# Remove outliers based on IQR (Interquartile Range) for better data representation
Q1 = car['Price'].quantile(0.25)
Q3 = car['Price'].quantile(0.75)
IQR = Q3 - Q1
car_clean = car[(car['Price'] >= (Q1 - 1.5 * IQR)) & (car['Price'] <= (Q3 + 1.5 * IQR))]

# Create the figure with two subplots: Swarmplot + Distribution
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Swarmplot for individual data points
sns.swarmplot(
    x='year',
    y='Price',
    data=car_clean,
    size=4,
    color='mediumslateblue',
    alpha=0.7,
    ax=ax1,
    label='Car Prices'
)

# Optional: Overlay mean price per year using pointplot
mean_prices = car_clean.groupby('year')['Price'].mean().reset_index()
sns.pointplot(
    x='year',
    y='Price',
    data=mean_prices,
    color='darkorange',
    markers='D',
    linestyles='--',
    scale=1.2,
    label='Mean Price',
    ax=ax1
)

# Add title and labels
ax1.set_title('Car Prices by Year (Cleaned Data)', fontsize=18, weight='bold')
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Price (in â‚¹)', fontsize=14)

# Rotate x-tick labels
for label in ax1.get_xticklabels():
    label.set_rotation(40)
    label.set_ha('right')
    label.set_fontsize(10)

# Highlighting outliers (with a different color)
outliers = car[~car.index.isin(car_clean.index)]
sns.swarmplot(
    x='year',
    y='Price',
    data=outliers,
    size=6,
    color='red',
    alpha=0.5,
    ax=ax1,
    label='Outliers'
)

# Add a vertical line to mark the mean price
mean_price_all = car['Price'].mean()
ax1.axhline(mean_price_all, color='gray', linestyle=':', linewidth=2, label=f"Overall Mean: â‚¹{mean_price_all:,.0f}")

# Second subplot for the distribution of car prices (Violin Plot)
sns.violinplot(
    x='year',
    y='Price',
    data=car_clean,
    inner="quart",
    ax=ax2,
    palette='viridis',
    hue='year',  # Set hue to 'year' to match the palette
    split=True,  # Split the plot based on 'year'
    legend=False  # Disable legend for now
)
ax2.set_title('Price Distribution by Year', fontsize=18, weight='bold')
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('Price (in â‚¹)', fontsize=14)

# Format y-axis with thousands separator
ax1.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'â‚¹{int(x):,}'))
ax2.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: f'â‚¹{int(x):,}'))

# Add grids
ax1.grid(axis='y', linestyle='--', alpha=0.5)
ax2.grid(axis='y', linestyle='--', alpha=0.5)

# Add legends manually
ax1.legend(loc='upper left', fontsize=10)
ax2.legend(loc='upper left', fontsize=10, labels=['Price Distribution', 'Outliers'])

# Tight layout to prevent overlap
plt.tight_layout()
plt.show()
```

## Extracting Training Data

```python
X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
X
```

```python
y
```

## Applying Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

## Creating an OneHotEncoder object to contain all the possible categories

```python
ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

##Creating a column transformer to transform categorical columns

```python
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

## Linear Regression Model

```python
lr=LinearRegression()
```

```python
lr
```

## Making a pipeline

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe
```

## fitting the model

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

## Checking R2 Score

```python
r2_score(y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

## The best model is found at a certain random state

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
